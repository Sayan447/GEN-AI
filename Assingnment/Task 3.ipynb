# Task 3: product pricing (dictionaries)

## create a dict price_dict where keys are product names and values are prices(int or floats). include at least 6 entries

```python
price_dict = {
    "Laptop": 75000,
    "Mouse": 500,
    "Keyboard": 1200,
    "Tablet": 25000,
    "Shoes": 3000,
    "Shirt": 1500
}
print(price_dict)
```

```python
# Write small code blocks to:
# Add a new product with price to price_dict.
                # - Update the price of an existing product.
                # - Remove a product by name (handle the case when the product doesn't exist).
```

```python
price_dict['Watch']  =  2500
print(price_dict)
```

```python
price_dict["Laptop"] = 72000
print(price_dict)
```

```python
product_name = "Mouse"

removed = price_dict.pop(product_name, None)

if removed is None:
    print("Product not found")
else:
    print(f"{product_name} removed successfully")

print(price_dict)
```

```python
# print the avg price of all products (use only dictionary operation and basic arithmetic)
```

# To calculate the average price of all products, you need:

## Sum of all prices

## Total number of products

## Then apply the formula:

## Average = Total Price / Number of Products


	​


```python
total_price = sum(price_dict.values())
total_products = len(price_dict)

avg_price = total_price / total_products


print("Average price: ", avg_price)
```

```python
# price_dict.values() → gets all prices

# sum() → adds all prices

# len(price_dict) → counts number of products

# total_price / total_products → gives average price
```

```python
max_product = max(price_dict, key=price_dict.get)
min_product = min(price_dict, key=price_dict.get)


print("Expensive product : " , max_product , '-' , price_dict[max_product])
print("Cheapest product : " , min_product, '-' , price_dict[min_product])
```
