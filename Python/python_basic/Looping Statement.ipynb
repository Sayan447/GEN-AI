# Looping Statement

## looping statement can be used for iterating over a series of numbers or through iterable

# or data structure or through a collection of data

```python
for item in range(1,20,2):
    print(item)
```

# by using range() we can represent the numbers in a decreamenting manner as well or in a decreasing order as well

```python
# print the number  from 50 to 1 with a gap of 1 value in a decreamenting manner
for i in range(50,1,-1):
    print(i)
```

```python
# print the number from 50 to 0 with gap of 5
for i in range(50,-1,-5):
    print(i)
    
```

```python
# In looping statement, we can iterate through iterabls such as string
name = "Amit Kumar"
for i in name:
    print(i)
```

# while loop

### while loop is being used for iterating over a series numbers when a condition is true

```python
i=0
while i<=50:
    print(i)
    i=i+1
    
```

```python
# we have to print the numbers from 0 to 50 as the even number
for i in range(51):
    if i%2 == 0:
        print(i)
```

```python
# print even number using while loop in btw 0 to 50
i = 0
while i<=50:
    if i%2==0:
        print(i)
        
    else: 
        pass
    i = i+1
```

```python
## print the number from 0 to 50 but once number=18 then terminate the loop or do not print
## either 18 or any number forward
for i in range(51):
    if i == 18:
        break
    else: print(i)
```

## while loop is being used for iterating over a series of numbers when a condition is True and we have to initialise a value as well while iterating

```python
# by making the use of while loop we need to print the number from 0 to 50
i=0
while i<=50:
    print(i)
    i=i+1
```

```python
# have to print the 0 to 50 even numbers
i=0
while i<=50:
    if i%2==0:
        print(i)
    else: pass
    i = i + 1
```

#### infinte loop ---> it is being used in a way where in the condition is not fullfilled

```python
# continue
# print even numbers
for i in range(20):
    if i%2 == 0:
        print(i)
    else:
        continue
    
```

```python
# print the number from 0 to 20 but once 17 is occuring skip it

for i in range(20):
  
    if i == 17:
        continue
    else: print(i)
```

### nested loops

```python
for i in range(6):
    for j in range(4):
        print(i,j)
```

```python
## printing the sum of first N natual numbers where natual number is 1,2,3

n = int(input("Enter the Value:- "))
sum = 0
for i in range(n+1):
    sum = sum + i
print(sum)
    
```
