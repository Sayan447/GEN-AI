```python
# csv library ---> Most of the time we are going to work with csv files so thats why knowledge 
# with csv library functionality matters
```

```python
import csv
```

```python
with open('example.txt' , mode='w', newline='') as file:
    writer=csv.writer(file)
    writer.writerow(['name' , 'age'])
    writer.writerow(['sonu' , 25])
    
    
    
# writerow() is a method of the Python csv module used to write one row of data into a CSV file.

# writerow() is used to write a single row of data into a CSV file, where each element of the list becomes a column separated by commas.
```

```python
with open('example.txt', mode='r') as file:
    # Creates a CSV reader object, It reads each line and splits data by comma
    reader = csv.reader(file)
    for row in reader:
        print(row)
        
        
# row[0] accesses the first element of the row list, which represents the first column of the CSV file.
```

```python
# datetime is used to work with date and time.

# timedelta is used to add or subtract time (days, hours, minutes, etc.).
from datetime import datetime, timedelta
time_current = datetime.now()
print(time_current)
```

```python
# When we get the JSON Data , we get in a JSON string format so by using json library we can convert our string data into dictionary format  for serializing our data
```

```python
data = {"Name" : "Amit" , "Age" : 25}
json_str = json.dumps(data)
print(json_str)
print(type(json_str))
```

```python
# dumps() means:
# dump string
# It converts Python object → JSON string.
# '{"Name": "Amit", "Age": 25}'



# dict
# ↓
# json.dumps()
# ↓
# JSON string
```

```python
# Why We Use json.dumps()

# To send data between systems like:

# Python → Web API

# Python → JavaScript

# Python → Database

# Python → File storage

# JSON is the most common data format in web applications.
```

```python
# json.dumps() converts a Python object (like a dictionary) into a JSON formatted string.
```

```python

json_dict = json.loads(json_str)
print(json_dict)
print(type(json_dict))



# This line converts a JSON string back into a Python dictionary using the Python json module.

# JSON string → Python object
```

```python
# Regular Expression

## Regular Expression is used to retrieve info based on a specific pattern or if the pattern matches
```

```python
import re
```

```python
text='Hello i am Sonu my password is sonu12345'
```

```python
pattern = r'\d+'
match=re.search(pattern,text)
print(match.group())

# \d+ is a regular expression pattern used to match one or more digits (numbers).
# It is used with the Python re module.
```

```python
# \d+ is a regular expression that matches one or more consecutive digits (0–9) in a string
```
