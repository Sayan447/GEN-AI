```python
# whether if the current year is a leap year or Not


year=int(input("Enter the year value: "))


if year%4 == 0:
    if year%100 == 0:
        if year%400 == 0:
            print('the current year is a leap year ')
        else: print("the current year is not Leap year")
    print("Year is Leap year")
    
else: print("year is not leap year")
```
