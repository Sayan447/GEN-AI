### shallow copy -> it's being used for creating a copy of original dictionary in a way such that the changes that we are going to make with original dict won't be getting reflected into our copied dict

```python
dict_1 = {'Name' : "amit" , "a" : 34 , "b" : 444}
dict_1
```

```python
dict_2 = dict_1
dict_2
```

```python
dict_2['a']
```

```python
dict_2['a'] = 99
dict_2['a']
```

```python
dict_2
```

```python
dict_1
```

```python
# shallow copy
dict_3 = dict_1.copy()
dict_3
```

```python
dict_1['a'] = 79
dict_1
```

```python
dict_3
```

# Nested dictionary

```python
nested_dict={
    'student_1' : {'Name' : 'Amit' , 'Mark' : 844 , 'Grade' : 'A'},
    'student_2' : {'Name' : 'Vivek' , 'Mark' : 975 , 'Grade' : 'A++'},
    'student_3' : {'Name' : 'Vivo' , 'Mark' : 792 , 'Grade' : 'B'},
}
nested_dict
```

```python
for Key,Value in nested_dict.items():
    print(f'key is :{Key} and Value is : {Value}')
    for key_1,val_1 in Value.items():
        print(f'inner key is : {key_1} and inner value is : {val_1}')
```

# dictionary comprehention

```python
com = {x**2 for x in range(11)}
com
```

```python
comp = {x:x**2 for x in range(11)}
comp
```

```python
even_dict = {x:x**2 for x in range(11) if x%2==0}
```

```python
even_dict
```

# Merged Dictionary

```python
dictionary_1 = {'a' : 1 , 'b' : 2, 'c' : 3}
dictionary_2 = {'h' : 9 , 'd' : 4, 'e' : 7}

Merged_dictionary = {**dictionary_1, **dictionary_2}
# ** → Unpacks key-value pairs from a dictionary.
Merged_dictionary
```
