## List Methods
1. append() -> insert the value at the end of the list
1. insert() -> insert the value at the specific index location of the list
1. remove() -> remove the value from our list that we are going to pass inside it
1. pop() -> remove the value from the list at the specific index location.
1. index() -> it provides the index of the value that we are going to pass inside it if that value is present inside that list

## Enumerate --> helps us with getting the index value along with the actual value in a combination we are storing inside the list 

```python
Numbers = [11,2,5,44,69,5,53,22]
```

```python
for index_value,value in enumerate(Numbers):
    print(index_value,value)
```

```python
length = len(Numbers)
print(length)
```

# List comprehension syntax

## [expression for item in iterable if condition]

### expresion --> what we want to print
### for item in iterable ---> looping statement
### if condition ---> consdition statement

```python
# I need to store values from 1 to 50 inside the list
```

```python



Numbers = [x for x in range(1,51)]
print(Numbers)
```

```python
# if we want only looking for those number which are even
even_items = [x**2 for x in range(1,19) if x%2==0]
even_items
```

<!-- Function in list comprehension as well -->

```python
list_itemss = ['amit' , 'apple' , 'banana' , 'car' , 'truck']
```

```python
length_list= [len(word) for word in list_itemss]
length_list
```

```python
pair = [[i,j] for i in list_itemss for j in length_list]
```

```python
pair
```

```python
lst_1=[1,3,5]
lst_2=[2,6,9]

paired_list=[i for i in lst_1]
paired_list1=[i for i in lst_2]
paired_list, paired_list1 #([1, 3, 5], [2, 6, 9])
# In Python, when you separate values with a comma, Python automatically packs them into a tuple.
# print(paired_list, paired_list1) # [1, 3, 5] [2, 6, 9]
```

```python
pair_lst = [[i,j] for i in lst_1 for j in lst_2]

print(pair_lst)
```
