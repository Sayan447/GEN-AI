### Map()

### map function applies a given fucntions operation to all items in an iterable or a list and return map object (an Iterator)

## map function is a particularly usefull for transforming the data or applying an operation on all items of the iterable

```python
# Squaring all items of list

def square(list_items):
    empty_list=[]
    for item in list_items:
        square_item = item **  2
        empty_list.append(square_item)
    print(empty_list)
    
square([2,4,8,10])
```

```python
# similar problem statement with a MAP() and lambda ------>
print(list(map(lambda x:x**2,[2,4,8,10])))
```

```python
# only the map()---->
def square(num):
    return num ** 2
# print(list(map(square([2,4,8,10]))))
# Here Python thinks you're calling square() with the whole list [2,4,8,10].
# map(function, iterable)

print(list(map(square, [2,4,8,10])))
```

## whenever you are going to pass the general function inside your map function as the parameter then you don't have to use parenthesis

```python
# Map fucntion with multiple Iterables
num1 = [1,4,5,8]
num2 = [8,5,6,7]
print(list(map(lambda x, y: x+y, num1,num2)))
```

```python
# map() to convert all string values items in an iterable into an integer
print(list(map(int ,['1' , '2' , '9' , '5' , '6' , '8' ])))
```

```python
emp_name = ["sonu" , 'sayan' , 'bubu']
print(list(map(str.upper,emp_name)))
```

```python
# Retrieve the values inside a dictionary using Map()
emp=[
    {'Name' : 'amit' , "Age" : 23 , 'qualification' : 'B.tech'},
    {'Name' : 'Vivek' , "Age" : 78 , 'qualification' : 'B.tech'},
    {'Name' : 'Sonu' , "Age" : 38 , 'qualification' : 'B.tech'},]


def age_retrieve(lst_item):
    return lst_item['Age']

print(list(map(age_retrieve, emp)))
c = list(map(age_retrieve, emp))
```

# Filter function

### filter function is going to retieve only those items which are meeting a certain conditions in an iterable

```python
# syntax of filter
# list(filter(func , iterable))
```

```python
# to retieve only those values which are even
def even_fun(value):
    if value%2==0:
        return True
lst = [1,2,4,5,6,3,9,38,58]
print(list(filter(even_fun , lst)))
```

```python
# solving the similar problem statement with a lambda function
print(list(filter(lambda x: x%2==0,lst)))
```

```python
# I want to retrieve the values which are greater than 5
print(list(filter(lambda x:x>5 , lst)))
```

```python
# I want value which are both grater than 5 and even
print(list(filter(lambda x: x%2==0 and x>5,lst)))
```

```python
# iterating over the dictionary items
emp = [
    {"Name" : 'amit' , 'Age' : 45, 'Qualification' : 'b.com'},
    {"Name" : 'dudun' , 'Age' : 75, 'Qualification' : 'M.S'},
    {"Name" : 'golu' , 'Age' : 59, 'Qualification' : 'M.com'},
    {"Name" : 'golu' , 'Age' : 25, 'Qualification' : 'M.com'},
]
```
