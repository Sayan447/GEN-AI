# Tuples

### it's kind of a data structure which can store the data in a ordered way similar to the list but tuples are immutable

```python
a=tuple()
type(a)
```

```python
tuple_values = (1,2,3,4,5)
print(tuple_values)
```

```python
tuple_values[1] = 24
```

```python
list_v = [1,2,3,4,5]
list_v[1]

```

```python
list_v[1] = 24
list_v
```

### tuples are also capable of storing the data of any data type

```python
tuple_data = (1,2,3,4,6,'Amit',4.7,True)
tuple_data
```

```python
tuple_data[::-1]
```

```python
tuple_data[1:7]
```
