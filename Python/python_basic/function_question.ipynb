```python
# to create function that converts temp from celcius to farenheit or viceversa
# if the temp is celcius then  c to f convertion is

"""1 deg celcius = 9/5 + 32
    45 deg C = 45 * 9/5 + 32 F
    
    1 F = 5/9 - 32 
    45 F = (45-32) * 5/9 c """
    
from math import ceil


def temp_change(temp , unit):
    if unit == 'C':
        temperature = temp * 9/5 +32
        print(f"the temp for {temp} deg celcius is F : ", str(temperature) + 'F' )
        
    elif unit == "F":
        temperature = (temp-32) * 5/9 
        print(f"the temp for {temp} deg F is C : ", str(round(temperature)) + ' C' )
    
    else: print("Invalid Input")
    



temp_change(1 , "F")
temp_change(1 , "C")
        
        
```

```python
## define a function that would behave as a Password Strength Checker
def password_strenght(password):
    if len(password) < 8:
        print(f'this is the Weak password {password} ')
    
    elif not any(char.islower() for char in password):
        print(f'this is the Weak password {password} ')
        
        
    elif not any(char.isupper() for char in password):
        print(f'this is the Weak password {password} ')
        
    elif not any(char.isdigit() for char in password):
        print(f'this is the Weak password {password} ')
    
    elif not any(char for char in '!,@,#,$,%,^,&,*,(,),_'):
        print(f'this is the Weak password {password} ')
        
    else:
        print(f'the entered password {password} is a strong password')
```

```python
password_strenght('Hello')
```

```python
password_strenght("Heloo125$")
```

```python
# check whether the current string is palindrome or not

def palindrome(string):
    string_value=string.lower().replace(" ","")

    if string_value == string_value[::-1]:
        print("Palindrome")
    else: print("Not palindrome")
    
palindrome('madam')
palindrome('AaBB aa')
```
