## Writing the Content into the file


## write()--> it overwrites the content into the text file

```python
#  write() --> helps us out in writing the content into the file
f=open('exapmle.txt','w')
f.write('Hello we are the Hunters')
f.close()
```

```python
# append to the end
f=open('exapmle.txt', 'a')
f.write('\nHello')
f.close()
```

## with --> it will be used for first of all opening the file as a object and there is no requirement for closing the file as well

```python
with open('exapmle.txt' , 'r') as f :
    content_data=f.read()
    print(content_data)
```

```python
# Reading the total lines , total words , total characters from a file

def count_lines(file):
    with open(file,'r') as files:
        lines = files.readlines()
        total_lines = len(lines)
        total_words = sum(len(line.split()) for line in lines)
        total_characters = sum(len(line) for line in lines)
    
    return total_lines,total_words,total_characters

file_path= "Total_data.txt"
total_lines , total_words , total_characters = count_lines(file_path)

print(total_lines , total_words , total_characters)
```

```python
# Write the content into a file and then after writing the content into the file we are
# going to read the content from that file as well --> seek()
```

#### w+ --> first of all as W is their so it states that first writing  operation will be done and then reading operation will be done

#### seek() --> it is going to help us out with reading the content after writing from the starting only

```python
# seek() moves the file pointer (cursor) to a specific position in the file.
# file.seek(offset, whence)

# offset → number of bytes to move
# whence → reference position (optional)

# | Value | Meaning                  |
# | ----- | ------------------------ |
# | 0     | From beginning (default) |
# | 1     | From current position    |
# | 2     | From end of file         |








with open('Total_data.txt' , 'w+') as write_read_file:
    write_read_file.write('\n We are looking into W+ operation')
    write_read_file.write('\n It means writing and then reading')
    write_read_file.seek(0)
    content_complete_data=write_read_file.read()
    print(content_complete_data)
    
    
```
