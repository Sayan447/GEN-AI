# OS library

## OS library is able to interact with our operating system and can let us work with multiple directories and file

```python
# getcwd() --> it is capable of providing us with the idea about the current working directory
```

```python
import os
```

```python
current_path=os.getcwd()
print('The current working directory is : ', current_path)
```

```python
# mkdir() ---> make directory --> It will be helping us out with creating the directory
```

```python
new_path="New_folder"
os.mkdir(new_path)
print(f"the new directiory {new_path} is creating successfully")
```

```python
# listdir() --> listing all of the directories present - list directories
# listdir() → returns a list of all files and folders inside a directory.
# '.' → means current directory (where your Python file is running).
All_items = os.listdir('.')
print(All_items)
```

```python
# joining multiple paths together
# os.path.join() --> it'll be joining multiple path together
dir_name="folder_data"
file_name = 'File_data'
full_path = os.path.join(dir_name,file_name)
print(f'the ful path is {full_path}')
```

```python
# I want to join multiple path with the current path
current_complete_path = os.path.join(os.getcwd(),dir_name,file_name)
print(f'the complete path is : {current_complete_path} ')
```

```python
# os.path.exists()
given_path = 'C:\Desktop\GEN AI\Python\python_basic\Dict.ipynb'

if os.path.exists(given_path):
    print('yes this path is exist')
else: print('this path is not exists')
```

```python
# os.path.isfile() --> this checks that whether if the current path is a file or not
# os.path.isdir() --> this checks that whether if the current path is a directory or not
```

```python
# path = "Total_data.txt"
path = "New_Folder"

if os.path.isfile(path):
    print('the current path belongs to a file')

elif os.path.isfile(path):
    print('the cuurent path belongs to a directory')

else: print('the current path is neighther a file or folder')
```
