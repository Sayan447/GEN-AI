# Recursive function

### by a recursive function we can say that a function is a going to call itself again and again

```python
## Factorial of Number
def factorial(num):
    if num==0 or num==1:
        print(f'the factorial of {num} is : 1')
    else: 
        factorial = 1
        for i in range(1,num+1):
            factorial = factorial * i
        print(f'the factorial of num is : {factorial}')
```

```python
number= int(input("enter the number : "))
factorial(number)
```

```python
def fact_recursive(num):
    if num==0 or num==1: 
        return 1
    else:
         return num * fact_recursive(num-1)
    
fact_recursive(5)
```

```python
def data(**user_data):
    for key,value in user_data.items():
        print(f'{key}:{value}')
    
```

```python
data(name="amit",age=24,gender="Male")
```

```python
# ** is used to work with keyword arguments (key=value pairs).
# It is mostly used in two places:

# ✅ A) In function definition → to collect keyword arguments
# ✅ B) In function call → to unpack a dictionary


# **data collects all keyword arguments
# It stores them in a dictionary
```

## we want to pass keyword value pairs as your input along with positional arguments as well

```python
def variable(*positional,**keyword):
    for i in positional:
        print(f'positional arguments ', i)
    for key,value in keyword.items():
        print(f'key is : {key} and value is :{value}')
```

## *args → Collects positional arguments - *args stores values in a tuple
## **kwargs → Collects keyword arguments - **kwargs stores values in a dictionary

```python
variable(1,2,3,4,5, name="Amit" , age=25, qualification="B.tech in cs")
```
