## Lambda Function


## lambda fucntion are the one liner functions defined by the lambda function

```python
# function which calculates the sqaure of the number
def square(num):
    return num**2

print(square(5))
```

```python
# using lambda function 
lambda_function=lambda x:x**2
```

```python
type(lambda_function)
```

```python
lambda_function(5)
```

```python
# lambda function to calculate the sum of two number



lambda_function1 = lambda x,y: x + y
lambda_function1(5,8)
```

```python
product = lambda a,b,c,d : a * b*c*d
print(product(4,8,9,3))
```
